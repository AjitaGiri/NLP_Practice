## Bag of Words

It's a technique used to convert text into numbers so that ML models can understand it. I t creates a list of unique words (vocabulary) in the dataset and then for each sentence it counts how many times each word appears.

### implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
 
documents=[
    'I love pasta',
    'Pasta is delicious in Italy',
    'I love pizza',
    'Pizza is yummy'
]

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(documents)
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out())
print("\n BoW Matrix:\n",X.toarray())
```

### pros
- easy to implement
- fixed size (row*column)
- work with ML
- fast & efficient

### cons
- sparsity
- OOV(OUT OF VOCABULARY)
- semantic meaning

## N-grams

```python
from sklearn.feature_extraction.text import CountVectorizer
 
documents=[
    'I love pasta',
    'Pasta is delicious in Italy',
    'I love pizza',
    'Pizza is yummy'
]

vectorizer=CountVectorizer(ngram_range=(2,2))
X=vectorizer.fit_transform(documents)
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out())
print("\n BoW Matrix:\n",X.toarray())
```
